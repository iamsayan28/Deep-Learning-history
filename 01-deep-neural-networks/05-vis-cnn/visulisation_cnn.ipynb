
# Striving for Simplicity: The All Convolutional Net (2014)

> Review notes of the above paper

- This paper introduces two important ideas: replacing max pooling with increased stride size and visualizing CNN layers.

- It proposes an architecture with only convolutional layers (no fully connected layers).

- It also replaces fully connected layers with global average pooling.

- They achieved good accuracy even after removing some traditional functions.


> Let's talk about the most important topic

> The First Three Layers of the CNN

- In the very beginning, the layers barely learn the features of images.
  
- But as you go deeper, the feature learning improves significantly.

> The 6th and 9th Layers

- At these layers, the network starts learning more complex features of the image.





> Around the 12th Layer (Pre-final Layer)

- These images show how the network focuses on different parts of an image of a cat.
  
- In the final convolutional layer, it seems to start learning the faces.





> How They Do It

- For each neuron, if the activation during the forward pass was negative, they set the gradient to zero during the backward pass.

- They pass only positive gradients back through the network, meaning only the neurons that had positive activations will be part of the gradient flow.



> Last Notes

- As we go deeper into the network, the CNN learns more and more about the input data.
