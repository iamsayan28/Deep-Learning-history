
# BERT: Bidirectional Encoder Representations from Transformers

This was published paper by *Devlin et al.*, presented at ACL 2019, introduced **BERT** (Bidirectional Encoder Representations from Transformers), a Transformer-based language representation model. BERT turn out to be one of most useful model in natural language processing by proposing pre-training of bidirectional representations from unlabeled text, enabling the model to condition on both left and right context in all layers. This bidirectional approach contrasts with traditional unidirectional language models which help in understanding the language more effifciently.



> Masked Language Modeling (MLM): 
Often likened to the Cloze task (*Taylor, 1953*), MLM involves randomly masking 15% of the input tokens in a sentence and training the model to predict these masked tokens. The final hidden states of the masked positions are passed through a softmax layer to predict the masked tokens.This approach allows the model to capture bidirectional dependencies in text.

> Next Sentence Prediction (NSP):**  
NSP addresses tasks that require understanding relationships between sentences, such as Question Answering (QA) and Natural Language Inference (NLI). For NSP, the model is trained on pairs of sentences: 50% consecutive pairs and 50% random pairs. The task is to classify whether the second sentence follows the first, helping BERT grasp sentence-level coherence.


BERT achieves good performance across various NLP tasks with minimal task-specific modifications. Fine-tuning involves adding a simple output layer on top of the pre-trained BERT model. This makes it highly versatile for tasks like: Sentiment analysis, text classification, queation answering and all. and the model turn out to be SOTA wihtout any arch changes.



> model configs

- 12 Transformer layers (blocks), 12 attention heads, and 110 million parameters.
- 24 Transformer layers (blocks), 16 attention heads, and 340 million parameters.
- Up to 512 tokens.  
- Word embedding dimensions (BERT Base): 768.  


BERT achieved unprecedented results across NLP benchmarks : 

- **GLUE score:** Improved to 80.5% (*7.7% absolute improvement*).
- **MultiNLI accuracy:** Reached 86.7% (*4.6% absolute improvement*).
- **SQuAD v1.1 Test F1:** Attained 93.2 (*1.5 points absolute improvement*).
- **SQuAD v2.0 Test F1:** Scored 83.1 (*5.1 points absolute improvement*).

These results demonstrated that **unsupervised pre-training** significantly enhances language understanding systems, even for low-resource tasks.
