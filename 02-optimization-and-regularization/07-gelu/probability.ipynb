> what is probability distributions ? 

A probability distribution is a list of all possible outcomes of a random variable along with their corresponding probability values.

`Example 1: Coin Toss`

- Consider the random experiment of tossing a coin. The possible outcomes are:
  - Heads (H)
  - Tails (T)

- The probability distribution for this experiment would list these outcomes along with their corresponding probabilities:
  -  Probability of Heads (H) = 1/2
  -  Probability of Tails (T) = 1/2

- So, the probability distribution for a coin toss is:
  - {H, T} with probabilities {1/2, 1/2}

### Code for two types of pdf(probability distribution function) : 

```python
# Discrete (PDF)
import matplotlib.pyplot as plt

# Values of the discrete random variable
x_values = [0, 1, 2, 3, 4]

# Probabilities associated with each value
probabilities = [0.1, 0.2, 0.4, 0.2, 0.1]

# Plotting the PDF
plt.stem(x_values, probabilities)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Probability Density Function (PDF) of Discrete Random Variable')
plt.xticks(x_values)
plt.ylim(0, max(probabilities) + 0.05)
plt.grid(True)
plt.show()
```

```python
#Continous (PDF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Generate values for the x-axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF values
pdf = norm.pdf(x, mu, sigma)

# Plotting the PDF
plt.plot(x, pdf, label=f'Normal Distribution\n$\mu={mu}$, $\sigma={sigma}$')
plt.xlabel('X')
plt.ylabel('Density')
plt.title('Probability Density Function (PDF) of a Continuous Random Variable')
plt.legend()
plt.grid(True)
plt.show()
```

`Prob Mass Function(PMF)`

- It is a mathematical function that describe the prob distribution of random descrete variable. 

- PMF of a random descrete variable assigns the a prob to each possible values. 

- PMF must satisfy the following cases :
  
  - prob assign to each values must be + or can be zero(obv).

  - sum of all prob shoudl be 1. (it sounds like very common rule).

- math, for a discrete random variable $X$ that can take values $x_1, x_2, \ldots, x_n$:

- The PMF is denoted by $P(X = x_i) = p_i$, where $p_i$ is the probability that $X$ equals $x_i$.

- The conditions for the PMF are:
  - $ p_i \geq 0 $ for all $ i $.
  - $ \sum_{i} p_i = 1 $.

- ***For example***, the PMF for the outcome $X$ of rolling the die is:

$$ 
P(X = x) = 
\begin{cases} 
\frac{1}{6} & \text{if } x \in \{1, 2, 3, 4, 5, 6\} \\
0 & \text{otherwise}
\end{cases} 
$$

- In this case, each outcome (1, 2, 3, 4, 5, 6) has an equal probability of $ \frac{1}{6} $, and the sum of these probabilities is 1.

`PMF Representation:`
- $ P(X = 1) = \frac{1}{6} $
- $ P(X = 2) = \frac{1}{6} $
- $ P(X = 3) = \frac{1}{6} $
- $ P(X = 4) = \frac{1}{6} $
- $ P(X = 5) = \frac{1}{6} $
- $ P(X = 6) = \frac{1}{6} $

- these probabilities add up to 1, satisfying the requirement that the sum of all probabilities for a discrete random variable must equal 1.

## code for PMF

```python
# Ex. Two dice sum

import random 
import pandas as pd 
L =[]

for i in range(1000):
    
    a = random.randint(1,6)
    b = random.randint(1,6)
    L.append(a + b)
```

```python
#plot 

s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()

s.plot(kind='bar')
plt.title("PMP of two dice rolling")
```

> Let's calculate the CDF of the PMF

- The cumulative distribution function (CDF) $F(x)$ describes the probability that a random variable $X$ will be found at a value less than or equal to $x$.

- math, $F(x) = P(X \leq x)$.

for a fair six-sided die, each outcome (1, 2, 3, 4, 5, 6) has an equal probability of $\frac{1}{6}$.

now, let's calculate the CDF step by step:

- The PMF gives us the probability of each outcome:
  - $P(X = 1) = \frac{1}{6}$
  - $P(X = 2) = \frac{1}{6}$
  - $P(X = 3) = \frac{1}{6}$
  - $P(X = 4) = \frac{1}{6}$
  - $P(X = 5) = \frac{1}{6}$
  - $P(X = 6) = \frac{1}{6}$

now, let's find the CDF, which is the sum of probabilities up to each value:

- For $x < 1$: $F(x) = 0$ (since $X$ cannot be less than 1)
- For $1 \leq x < 2$: $F(x) = P(X \leq 1) = P(X = 1) = \frac{1}{6}$
- For $2 \leq x < 3$: $F(x) = P(X \leq 2) = P(X = 1) + P(X = 2) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6}$
- For $3 \leq x < 4$: $F(x) = P(X \leq 3) = P(X = 1) + P(X = 2) + P(X = 3) = \frac{3}{6}$
- For $4 \leq x < 5$: $F(x) = P(X \leq 4) = P(X = 1) + P(X = 2) + P(X = 3) + P(X = 4) = \frac{4}{6}$
- For $5 \leq x < 6$: $F(x) = P(X \leq 5) = P(X = 1) + P(X = 2) + P(X = 3) + P(X = 4) + P(X = 5) = \frac{5}{6}$
- For $x \geq 6$: $F(x) = P(X \leq 6) = P(X = 1) + P(X = 2) + P(X = 3) + P(X = 4) + P(X = 5) + P(X = 6) = 1$

so, the CDF for this fair six-sided die is:

\[ 
F(x) = 
\begin{cases} 
0 & \text{if } x < 1 \\
\frac{1}{6} & \text{if } 1 \leq x < 2 \\
\frac{2}{6} & \text{if } 2 \leq x < 3 \\
\frac{3}{6} & \text{if } 3 \leq x < 4 \\
\frac{4}{6} & \text{if } 4 \leq x < 5 \\
\frac{5}{6} & \text{if } 5 \leq x < 6 \\
1 & \text{if } x \geq 6
\end{cases}
\]

this means:
- If you want to know the probability of rolling a 3 or less, you add up the probabilities for 1, 2, and 3.
- The CDF gives you the total probability up to any value $x$. 

for example:
- The probability of rolling 2 or less (CDF at 2) is $\frac{2}{6}$.
- The probability of rolling 4 or less (CDF at 4) is $\frac{4}{6}$.
- And so on, until the total probability reaches 1 for all values 6 or higher.

## Code for  CDF of the PMF

```python

s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()

s = np.cumsum(s)

s.plot(kind='bar')
plt.title("Cumulative Sum of PDF")
```

> estimation methods os estimation of pdf 

`Parametric Methods for PDF`

`Parametric methods assume that the data follows a specific distribution with a set of parameters.`

- Example: If we assume that people’s heights follow a normal distribution, we can describe the PDF using two parameters: the mean (average height) and the standard deviation (how spread out the heights are).

`How it works:`
1. choose a distribution (e.g., normal, exponential).
2. estimate the parameters of the chosen distribution from the data.
3. use these parameters to define the PDF**.


- simple and efficient if the chosen model is correct.
- requires fewer data points.
- If the chosen distribution doesn’t fit the data well, the estimates can be misleading.

> KDE (Kernel Density Estimation) -> non-paramatric estimation method

- here we don't assume any that our is distribution normal or poisson etc.

- the main advantages is that it does not require the assumption of a specific distribution, which allows for more flexible and accurate estimation in situations where the underlying distribution is unknown or complex.

- computationally intensive and may require more data to achieve accurate estimates compared to parametric methods.
